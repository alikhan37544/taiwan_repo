```python
import tensorflow as tf
import numpy as np
from tensorflow.keras import layers, models, applications
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import os

# Split the data into features and labels
image_ids = df.iloc[:, 0].values  # First column contains image identifiers
labels = df.iloc[:, 1:].values.astype(np.float32)  # Convert to float32

# Split into train/test sets
train_ids, test_ids, y_train, y_test = train_test_split(
    image_ids, labels, test_size=0.2, random_state=42
)

# Define image dimensions
IMG_HEIGHT = 224
IMG_WIDTH = 224
CHANNELS = 3

# Define improved model with attention mechanism
def create_improved_model(input_shape=(224, 224, 3), num_classes=17):
    # Data augmentation layers
    data_augmentation = tf.keras.Sequential([
        layers.RandomFlip("horizontal_and_vertical"),
        layers.RandomRotation(0.2),
        layers.RandomZoom(0.2),
        layers.RandomContrast(0.2),
        layers.RandomTranslation(0.1, 0.1),
        layers.RandomBrightness(0.2),
    ])
    
    # Use EfficientNetV2S - more powerful than the original model
    base_model = applications.EfficientNetV2S(
        weights='imagenet',
        include_top=False,
        input_shape=input_shape
    )
    base_model.trainable = False
    
    inputs = layers.Input(shape=input_shape)
    x = data_augmentation(inputs)
    x = base_model(x)
    
    # Add attention mechanism
    attention = layers.Conv2D(1, (1, 1), activation='sigmoid')(x)
    x = layers.Multiply()([x, attention])
    
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.BatchNormalization()(x)
    
    # Add deeper and regularized classification layers
    x = layers.Dense(512, activation='relu')(x)
    x = layers.Dropout(0.5)(x)
    x = layers.Dense(256, activation='relu')(x)
    x = layers.Dropout(0.3)(x)
    
    # Output layer for multi-label classification
    outputs = layers.Dense(num_classes, activation='sigmoid')(x)
    
    model = tf.keras.Model(inputs, outputs)
    
    # Calculate class weights for handling imbalance
    class_weights = {}
    for i, cls_name in enumerate(df.columns[1:]):
        # Get count from class_count DataFrame
        count = class_count[class_count['class'] == cls_name]['Count'].values[0]
        # Inverse frequency weighting
        class_weights[i] = 1.0 / (count + 10)  # Adding 10 to avoid extreme weights
    
    # Normalize weights
    total = sum(class_weights.values())
    for i in class_weights:
        class_weights[i] = class_weights[i] * len(class_weights) / total
    
    # Compile model with standard binary crossentropy
    model.compile(
        optimizer=Adam(learning_rate=0.001),
        loss='binary_crossentropy',
        metrics=[
            'accuracy',
            tf.keras.metrics.AUC(name='auc'),
            tf.keras.metrics.Precision(name='precision'),
            tf.keras.metrics.Recall(name='recall')
        ]
    )
    
    return model, class_weights

# Create model and get class weights
model, class_weights = create_improved_model()

# Define callbacks for training
callbacks = [
    EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=1e-6),
    ModelCheckpoint('best_model.h5', monitor='val_loss', save_best_only=True)
]

# Create a function to load images from file paths
def load_and_preprocess_images(image_ids, img_dir='images', img_height=224, img_width=224):
    images = []
    for img_id in image_ids:
        img_path = os.path.join(img_dir, f"{img_id}.jpg")
        try:
            img = tf.keras.preprocessing.image.load_img(
                img_path, target_size=(img_height, img_width)
            )
            img_array = tf.keras.preprocessing.image.img_to_array(img)
            img_array = img_array / 255.0  # Normalize
        except Exception as e:
            print(f"Error loading {img_id}: {e}")
            # Fallback if image can't be loaded
            img_array = np.zeros((img_height, img_width, 3))
        
        images.append(img_array)
    
    return np.array(images)

# Training execution
# Step 1: Load all images (adjust img_dir to your actual image directory)
print("Loading training images...")
X_train = load_and_preprocess_images(train_ids, img_dir='images')
print("Loading test images...")
X_test = load_and_preprocess_images(test_ids, img_dir='images')

# Training phase 1: Training with frozen base model
print("Phase 1: Training with frozen base model")
history1 = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=30,
    batch_size=32,
    callbacks=callbacks,
    class_weight=class_weights,  # Apply class weights here
    verbose=1
)

# Plot training history
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history1.history['loss'], label='Train Loss')
plt.plot(history1.history['val_loss'], label='Validation Loss')
plt.title('Loss')
plt.xlabel('Epoch')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history1.history['accuracy'], label='Train Accuracy')
plt.plot(history1.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy')
plt.xlabel('Epoch')
plt.legend()
plt.tight_layout()
plt.show()

# Training phase 2: Fine-tuning
print("Phase 2: Fine-tuning the model")
# Get the base model
base_model = None
for layer in model.layers:
    if isinstance(layer, tf.keras.Model):
        base_model = layer
        break

if base_model:
    base_model.trainable = True
    
    # Freeze early layers, unfreeze later layers
    for layer in base_model.layers[:-30]:
        layer.trainable = False

    # Lower learning rate for fine-tuning
    model.compile(
        optimizer=Adam(learning_rate=1e-5),
        loss='binary_crossentropy',
        metrics=[
            'accuracy',
            tf.keras.metrics.AUC(name='auc'),
            tf.keras.metrics.Precision(name='precision'),
            tf.keras.metrics.Recall(name='recall')
        ]
    )

    # Continue training with fine-tuning
    history2 = model.fit(
        X_train, y_train,
        validation_data=(X_test, y_test),
        epochs=20,
        batch_size=16,  # Smaller batch size for fine-tuning
        callbacks=callbacks,
        class_weight=class_weights,  # Apply class weights here
        verbose=1
    )
    
    # Plot fine-tuning history
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history2.history['loss'], label='Train Loss')
    plt.plot(history2.history['val_loss'], label='Validation Loss')
    plt.title('Fine-tuning Loss')
    plt.xlabel('Epoch')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history2.history['accuracy'], label='Train Accuracy')
    plt.plot(history2.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Fine-tuning Accuracy')
    plt.xlabel('Epoch')
    plt.legend()
    plt.tight_layout()
    plt.show()

# Evaluate the model
print("Evaluating model on test data...")
test_metrics = model.evaluate(X_test, y_test)
print("\nTest metrics:")
for i, metric_name in enumerate(model.metrics_names):
    print(f"{metric_name}: {test_metrics[i]}")

# Generate predictions and compute per-class metrics
predictions = model.predict(X_test)
predictions_binary = (predictions > 0.5).astype(int)

# Calculate per-class performance
from sklearn.metrics import precision_score, recall_score, f1_score

# Calculate metrics for each class
class_names = df.columns[1:].tolist()
class_precision = precision_score(y_test, predictions_binary, average=None)
class_recall = recall_score(y_test, predictions_binary, average=None)
class_f1 = f1_score(y_test, predictions_binary, average=None)

# Display per-class performance
print("\nPer-class performance:")
print("Class\t\tPrecision\tRecall\t\tF1-Score")
for i, class_name in enumerate(class_names):
    print(f"{class_name:<15}\t{class_precision[i]:.4f}\t\t{class_recall[i]:.4f}\t\t{class_f1[i]:.4f}")

# Save the model
model.save('land_cover_classification_model.h5')
print("Model saved to 'land_cover_classification_model.h5'")

# Visualize some example predictions
def plot_predictions(X, y_true, y_pred, class_names, num_samples=5):
    indices = np.random.choice(len(X), num_samples, replace=False)
    
    plt.figure(figsize=(20, num_samples * 4))
    for i, idx in enumerate(indices):
        plt.subplot(num_samples, 2, 2*i+1)
        plt.imshow(X[idx])
        plt.title("Image")
        plt.axis('off')
        
        plt.subplot(num_samples, 2, 2*i+2)
        true_classes = [class_names[j] for j in range(len(class_names)) if y_true[idx][j] > 0.5]
        pred_classes = [class_names[j] for j in range(len(class_names)) if y_pred[idx][j] > 0.5]
        
        plt.barh(range(len(class_names)), y_pred[idx], color='skyblue')
        plt.yticks(range(len(class_names)), class_names)
        plt.xlim(0, 1)
        plt.title(f"True: {', '.join(true_classes)}\nPredicted: {', '.join(pred_classes)}")
    
    plt.tight_layout()
    plt.show()

# Visualize some predictions
plot_predictions(X_test, y_test, predictions, class_names, num_samples=5)
```
